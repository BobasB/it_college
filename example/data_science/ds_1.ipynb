```python
import random
```

```python
def make_trunk(colors, dist, total_in_trunk):
    total_dist = sum(dist)
    
    trunk = []
    for (color, count) in zip(colors, dist):
        num = count / total_dist * total_in_trunk
        trunk.extend([color] * int(num))
    random.shuffle(trunk)
    
    return trunk

def sample_n(trunk, n):
    sample = []
    for i in range(n):
        sample.append(trunk.pop())
    return sample
```

```python
colors = ['chartreuse', 'green', 'blue', 'red',
          'purple', 'pink', 'grey', 'black', 'white', 'brown']
dist1 = [20, 11, 5, 22, 3, 50, 1,  1, 12, 19]
dist2 = [20, 11, 5, 1,  3, 50, 20, 1, 20, 19]
#                   ^^         ^^     ^^
```

```python
trunk1 = make_trunk(colors, dist1, 1e7)
```

```python
import numpy as np
import matplotlib.pyplot as plt

#   
base_price = 1000

#  60   ( )
time_intervals = np.arange(0, 60, 1)  #  0  59 ,  1 

#    
price_fluctuations_1 = base_price + np.random.normal(0, 5, size=time_intervals.size)
price_fluctuations_2 = base_price + np.random.normal(0, 5, size=time_intervals.size)

#   
plt.plot(time_intervals, price_fluctuations_1, label="  1")
plt.plot(time_intervals, price_fluctuations_2, label=" 2")
plt.xlabel(" ()")
plt.ylabel(" (USD)")
plt.title("    ")
plt.legend()
plt.grid(True)
plt.show()
```
