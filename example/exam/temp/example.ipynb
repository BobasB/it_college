### Завдання
Створити клас `ProductCalculator`, який:
- Приймає в конструкторі список чисел
- Має метод `calculate_product()`, який обчислює добуток всіх елементів списку
- Має property `product`, яка повертає добуток всіх чисел
- Створити у циклі 3-5 об'єктів з різними списками та вивести результати

## АІ промпт
Ти є АІ помічник у написанні коду на Python. Твоє завдання допомогти користувачу виконати завдання, дотримуючись вимог, наведених вище.

Напиши код для Завдання у наступній комірці коду. Для коментарів, пояснень або документації у коді використовуй українську мову. Також текст у функції print має бути українською мовою.

```python
class ProductCalculator:
    """Клас для обчислення добутку елементів списку"""
    
    def __init__(self, numbers):
        """
        Ініціалізація об'єкта
        
        Args:
            numbers: список чисел
        """
        self.numbers = numbers
        self._product = None
    
    def calculate_product(self):
        """Обчислює добуток всіх елементів списку"""
        result = 1
        for num in self.numbers:
            result *= num
        self._product = result
        return result
    
    @property
    def product(self):
        """Property для отримання добутку"""
        if self._product is None:
            self.calculate_product()
        return self._product
    
    def __str__(self):
        """Повертає ім'я та прізвище студента"""
        return "Ім'я Прізвище Студента"


# Створення тестових даних
test_data = [
    [2, 3, 4],
    [5, 10, 2],
    [1, 2, 3, 4, 5],
    [-2, 3, 4],
    [7, 8]
]

# Створення об'єктів у циклі та виведення результатів
print("Демонстрація роботи класу ProductCalculator:\n")
for i, data in enumerate(test_data, 1):
    calc = ProductCalculator(data)
    print(f"Об'єкт {i}:")
    print(f"  Студент: {calc}")
    print(f"  Список чисел: {calc.numbers}")
    print(f"  Добуток елементів: {calc.product}")
    print()
```
