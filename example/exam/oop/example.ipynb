### Завдання
Створити клас `ProductCalculator`, який:
- Приймає в конструкторі список чисел
- Має метод `calculate_product()`, який обчислює добуток всіх елементів списку
- Має property `product`, яка повертає добуток всіх чисел

### Описи завдання
Напиши код для Завдання у наступній комірці коду з типом Python. Для коментарів, пояснень або документації у коді використовуй українську мову. Також, текст у функції print має бути українською мовою.

```python
class ProductCalculator:
    """Клас для обчислення добутку елементів списку"""
    
    def __init__(self, numbers, student_name="Ім'я Студента", group="Група"):
        """
        Ініціалізація калькулятора
        :param numbers: список чисел
        :param student_name: ім'я студента
        :param group: група студента
        """
        self.numbers = numbers
        self.student_name = student_name
        self.group = group
        self._product = None
    
    def calculate_product(self):
        """Метод для обчислення добутку всіх елементів списку"""
        self._product = 1
        for num in self.numbers:
            self._product *= num
        return self._product
    
    @property
    def product(self):
        """Property для отримання добутку"""
        if self._product is None:
            self.calculate_product()
        return self._product
    
    def __str__(self):
        """Метод для виведення інформації про студента"""
        return f"Студент: {self.student_name}, Група: {self.group}"


# Створення об'єктів класу з різними даними
calc1 = ProductCalculator([1, 2, 3, 4, 5], "Іван Петренко", "ІТ-21")
calc2 = ProductCalculator([10, 5, 2], "Марія Коваленко", "ІТ-22")
calc3 = ProductCalculator([-2, 3, 4], "Олександр Сидоренко", "ІТ-21")

# Демонстрація роботи класу
print("=== Демонстрація роботи класу ProductCalculator ===\n")

print(calc1)
print(f"Список чисел: {calc1.numbers}")
print(f"Добуток елементів: {calc1.product}")
print()

print(calc2)
print(f"Список чисел: {calc2.numbers}")
print(f"Добуток елементів (через метод): {calc2.calculate_product()}")
print()

print(calc3)
print(f"Список чисел: {calc3.numbers}")
print(f"Добуток елементів: {calc3.product}")
```
