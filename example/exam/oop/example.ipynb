### Описи завдання
Напиши код для Завдання у наступній комірці коду з типом Python. Для коментарів, пояснень або документації у коді використовуй українську мову. Також, текст у функції print має бути українською мовою.

```python
# Клас для обчислення добутку чисел
class ProductCalculator:
    def __init__(self, numbers, student_name="Студент", group="Група"):
        """
        Конструктор класу
        :param numbers: список чисел для обчислення добутку
        :param student_name: ім'я студента
        :param group: група студента
        """
        self.numbers = numbers
        self.student_name = student_name
        self.group = group
        self._product = None
    
    def calculate_product(self):
        """Метод для обчислення добутку всіх елементів списку"""
        result = 1
        for num in self.numbers:
            result *= num
        self._product = result
        return self._product
    
    @property
    def product(self):
        """Property для отримання добутку"""
        if self._product is None:
            self.calculate_product()
        return self._product
    
    def __str__(self):
        """Метод для виведення інформації про студента"""
        return f"Студент: {self.student_name}, Група: {self.group}"


# Створення 3 об'єктів класу з різними вхідними даними
calculators = [
    ProductCalculator([2, 3, 4], "Іван Петренко", "ІТ-101"),
    ProductCalculator([1, 5, 2, 3], "Марія Коваленко", "ІТ-102"),
    ProductCalculator([10, 2, 5], "Павло Сидоренко", "ІТ-101")
]

# Демонстрація роботи класу
for i, calc in enumerate(calculators, 1):
    print(f"\n--- Об'єкт {i} ---")
    print(calc)
    print(f"Список чисел: {calc.numbers}")
    print(f"Добуток: {calc.product}")
```
