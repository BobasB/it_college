### Властивості
- property та setter

```python
class Person:
    def __init__(self, age):
            self.__age = 0
            self.age = age #коли створюємо обєкт
    @property
    def age(self):
            return self.__age
    #назва setter має бути така сама як property
    @age.setter
    def age(self, age):
            if age > 0:
                    self.__age = age
p = Person(35)
print("після створення",p.age)
p.age = 25 # проперті без сетера неможна переприсвоювати
print("після setter", p.age)
```
